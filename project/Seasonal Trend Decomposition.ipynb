```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
plt.style.use("seaborn")
df = pd.read_csv("multiTimeline(1).csv")
df.columns = ["'computer science' Interest Over time"]
df = df[1:]
df.index = df.reset_index()['index'].apply(lambda z: datetime.datetime.strptime(z, '%Y-%m'))
df[df.columns[0]]= df[df.columns[0]].apply(int)
plt.figure(figsize=(12,5))
p = plt.plot(df)
t = plt.title(df.columns[0])
plt.ylabel("Interest on Google")
plt.xlabel("Date")
```
